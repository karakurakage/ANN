```python
# funcao normatiza dados
def Normatiza(x,strings):
    for i in strings:
        max_x=x[i].max()
        min_x=x[i].min()
        x_M=(max_x+min_x)/2
        x[i]=(x[i]-x_M)/(max_x-x_M)
    return x
```

```python
# funcao normatiza dados
def Original(x,xcalc,strings):
    k=0
    for i in strings:
        max_x=x[i].max()
        min_x=x[i].min()
        x_M=(max_x+min_x)/2
        xcalc[k]=xcalc[k]*(max_x-x_M)+x_M
        k=k+1
    return xcalc
```

```python
def f(Time, TC):
    return 3386.1538+290.0481*Time+250.7966*TC

x = np.linspace(-2, 2, 30)
y = np.linspace(-2, 2, 30)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)

fig = plt.figure()
ax = plt.axes(projection='3d')
ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, Z, cmap=plt.cm.jet, rstride=1, cstride=1, linewidth=0)
ax.set_xlabel('Time')
ax.set_ylabel('TC')
ax.set_zlabel('MW');
```

```python
def MW_calc(Time, TC):
    return 3386.1538+290.0481*Time+250.7966*TC
def MI_calc(Time, TC):
    return 69.5222-13.1952*TC**2
def Y_calc(Time, TC):
    return 79.9400+1.4070*Time+0.7285*TC-2.7521*Time**2-2.0021*TC**2
def Desirability(y,U,T,L):
    if y<L:
        d=0
    if (y>=L)&(y<=T):
        d=(y-L)/(T-L)
    if (y>=T)&(y<+U):
        d=(U-y)/(U-T)
    if (y>U):
        d=0
    return d
```

```python
def function(x):
    Time=x[0]
    TC=x[1]
    MW=MW_calc(Time, TC)
    MI=MI_calc(Time, TC)
    Y=Y_calc(Time, TC)
    U=3890
    T=3400
    L=2940
    d1=Desirability(MW,U,T,L)
    U=68
    T=65
    L=62
    d2=Desirability(MI,U,T,L)
    U=80
    T=77.5
    L=75.6
    d3=Desirability(Y,U,T,L)
    fun = (-1)*(d1*d2*d3)**(1/3)
    return fun
```

```python
bounds = [(-2, 2),(-2, 2)]
ret = optimize.differential_evolution(function, bounds,
                                      init="latinhypercube", # obs: init="random"
                                      maxiter=200, popsize=40,recombination=0.5)
```

```python
print("global minimum: x = ",ret.x, "f(x0) = ",ret.fun)
```
